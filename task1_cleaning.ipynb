```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Load dataset
df=pd.read_csv("customer_personality_analysis.csv")
```

```python
# data inspection
# first 5 rows of the dataset
df.head()
```

```python
#basic information about the dataset
df.info()
```

```python
#statistical summary of the dataset
df.describe()
```

```python
# cloumn names of the dataset
df.columns
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# 'income' column has 24 missing values in this dataset... so we are filling with median (for better than mean for skewed data)
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
# verifying if the missing values are filled
print(df.isnull().sum())
```

```python
# remove duplicates
print("Before:", df.shape)

df.drop_duplicates(inplace=True)

print("After:", df.shape)
```

```python
# fixing data types
# converting 'Dt_Customer' column to datetime format
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'], dayfirst=True)

#verifying the data type of 'Dt_Customer' column
print(df['Dt_Customer'].dtype)
```

```python
print(df['Marital_Status'].unique())
```

```python
# Fix inconsistent Marital_Status values
df['Marital_Status']=df['Marital_Status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
})

print(df['Marital_Status'].unique())
```

```python
# standardize text column 
df["Education"]=df["Education"].str.strip().str.title()

df["Marital_Status"]=df["Marital_Status"].str.strip().str.title()
```

```python
# create age column 
current_year=2026

df['age']=current_year-df['Year_Birth']
```

```python
df[['Year_Birth','age']].head()
```

```python
# check data types

df.dtypes
```

```python
# drop unnecessary columns: 
df.drop(['Z_CostContact','Z_Revenue'], axis=1, inplace=True) # this constant columns -no analytical value
```

```python
# save the cleaned dataset to a new CSV file
df.to_csv("customer_cleaned_data.csv", index=False)

print("Cleaned Dataset Saved!")
```
